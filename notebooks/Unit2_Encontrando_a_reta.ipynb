# Exemplo 01

## Etapa 01
Importar as bibliotecas de computação científica.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Etapa 02
Definir um conjunto de pontos e plotar o gráfico de dispersão.

```python
x = np.array([1, 4, 5, 8, 10])
y = np.array([3, 9, 11, 17, 21])

print('x:', x)
print('y:', y)
```

```python
plt.scatter(x, y, marker='o', color='b')
plt.xlabel('entrada - x')
plt.ylabel('saída - y')
plt.title('Conjunto de dados')
plt.grid()
plt.show()
```

## Etapa 03
Encontrar a reta que se ajusta aos pontos - linha de tendência.

```python
z = np.polyfit(x, y, 1)
print('z', z)
```

```python
z2 = np.polynomial.polynomial.Polynomial.fit(x, y, 1, [])
print(z2)
```

```python
p = np.poly1d(z)
print('Equação da reta que se ajusta aos pontos:\n', p)
```

```python
# Plotando a reta
plt.scatter(x, y, marker='o', color='b')
plt.plot(x, p(x), 'r--')
plt.xlabel('entrada - x')
plt.ylabel('saída - y')
plt.title('Exemplo de ML supervisionado')
plt.grid()
plt.show()
```

# Exemplo 02

## Etapa 01
Definir um conjunto de pontos e plotar o gráfico de dispersão.

```python
x = np.array([1, 4, 5, 8, 10])
y = np.array([3, 7, 11, 15, 22])

plt.scatter(x, y, marker='o', color='b')
plt.xlabel('entrada - x')
plt.ylabel('saída - y')
plt.grid()
plt.show()
```

## Etapa 02
Encontrar a reta que se ajusta aos pontos - linha de tendência.

```python
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
print('Equação da reta que se ajusta aos pontos: ', p)

# Plotando a reta
plt.scatter(x, y, marker='o', color='b')
plt.plot(x,p(x),'r--')
plt.xlabel('entrada - x')
plt.ylabel('saída - y')
plt.title("Exemplo de ML Supervisionado")
plt.grid()
plt.show()
```

Plotando um polinômio

```python
z = np.polyfit(x, y, 9)
p = np.poly1d(z)
print('Equação do polinomio que se ajusta aos pontos: \n', p)

# Plotando o polinomio
plt.scatter(x, y, marker='o', color='b')
t = np.linspace(0, 10, 200)
plt.plot(t, p(t), 'r--')
plt.grid()
plt.show()
```
