#Clusterização das espécies da flor Iris - abordagem não supervisionada.

##Etapa 01
Importar as bibliotecas de computação científica e aprendizado de máquina.

```python
#Bibliotecas de aprendizado de máquina
from sklearn import datasets
from sklearn.cluster import KMeans

#Bibliotecas de computação científica
import numpy as np
import matplotlib.pyplot as plt
```

##Etapa 02
Importar a base de dados.


```python
iris_dataset = datasets.load_iris()
X = iris_dataset.data
print(X.shape)
```

##Etapa 03
Definir e treinar o algoritmo de clusterização.

```python
#Definicão do algoritmo de clusterização
kmeans = KMeans(n_clusters=99)
wcss = []

#Treinamento do algoritmo
for k in range(1,15):
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

#Plotagem do gráfico do cotovelo
plt.figure(figsize=(20,8))
plt.title('Gráfico do cotovelo para clusterização das espécies de flores', fontsize=18)
plt.plot(range(1,15), wcss, '-o')
plt.grid(True)
plt.xlabel('Quantidade de clusters', fontsize=14)
plt.ylabel('Inercia', fontsize=14)
plt.xticks(range(1,15))
plt.show()
```

##Etapa 04
Clusterização das espécies de flores com diferentes K.

```python
plt.figure(figsize=(24,4))

plt.suptitle("Clusterização das espécies de flores com diferentes K", fontsize=20)

#Para K = 1
plt.subplot(1,6,1)
plt.title("K = 1",fontsize=16)
plt.xlabel("Comprimento da pétala (cm)")
plt.ylabel("Largura da pétala (cm)")
plt.scatter(X[:,2],X[:,3])

#Para K = 2
plt.subplot(1,6,2)
plt.title("K = 2",fontsize=16)
plt.xlabel("Comprimento da pétala (cm)")
kmeans = KMeans(n_clusters=2)
R= kmeans.fit_predict(X)
plt.scatter(X[:,2][R == 0],X[:,3][R == 0])
plt.scatter(X[:,2][R == 1],X[:,3][R == 1])

#Para K = 3
plt.subplot(1,6,3)
plt.title("K = 3",fontsize=16)
plt.xlabel("Comprimento da pétala (cm)")
kmeans = KMeans(n_clusters=3)
R= kmeans.fit_predict(X)
plt.scatter(X[:,2][R == 0],X[:,3][R == 0])
plt.scatter(X[:,2][R == 1],X[:,3][R == 1])
plt.scatter(X[:,2][R == 2],X[:,3][R == 2])

#Para K = 4
plt.subplot(1,6,4)
plt.title("K = 4",fontsize=16)
plt.xlabel("Comprimento da pétala (cm)")
kmeans = KMeans(n_clusters=4)
R= kmeans.fit_predict(X)
plt.scatter(X[:,2][R == 0],X[:,3][R == 0])
plt.scatter(X[:,2][R == 1],X[:,3][R == 1])
plt.scatter(X[:,2][R == 2],X[:,3][R == 2])
plt.scatter(X[:,2][R == 3],X[:,3][R == 3])

#Para K = 5
plt.subplot(1,6,5)
plt.title("K = 5",fontsize=16)
plt.xlabel("Comprimento da pétala (cm)")
kmeans = KMeans(n_clusters=5)
R= kmeans.fit_predict(X)
plt.scatter(X[:,2][R == 0],X[:,3][R == 0])
plt.scatter(X[:,2][R == 1],X[:,3][R == 1])
plt.scatter(X[:,2][R == 2],X[:,3][R == 2])
plt.scatter(X[:,2][R == 3],X[:,3][R == 3])
plt.scatter(X[:,2][R == 4],X[:,3][R == 4])

#Rótulos Originais
plt.subplot(1,6,6)
plt.title("Rótulos originais",fontsize=16)
plt.xlabel("Comprimento da pétala (cm)")
plt.scatter(X[0:49,2], X[0:49,3])
plt.scatter(X[50:99,2], X[50:99,3])
plt.scatter(X[100:149,2], X[100:149,3])
plt.subplots_adjust(top=0.8)
plt.show()
```
