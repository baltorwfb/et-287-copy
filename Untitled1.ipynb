**Modelo Malthusiano**

Um modelo bastante simples para descrever a dinâmica populacional pode ser obtida se admitimos que tanto a taxa de natalidade como a de mortalidade sejam constantes ao longo do tempo:

$$\alpha = \dfrac{P\left( t_{t+1} \right) - P \left(t_i \right)}{P \left(t_i \right)} = n - m$$

Nesse modelo é considerado que "a variação populacional é proporcional à própria população" (modelo malthusiano). Matematicamente podemos escrever:

$$\dfrac{d P(t)}{dt} = \alpha P(t)$$

Esta equação pode ser reescrita como:

$$\dfrac{dP}{P} = \alpha dt$$

Duas diferenciais iguais têm a mesma integral, a menos de uma constante arbitrária, isto é:

$$\int \dfrac{dP}{P} =  \int \alpha dt + k$$

Realizando a integração, temos:

$$ln P(t) = \alpha t + k \implies$$

$$P(t) = e^{t} \, e^{\alpha t}$$

 **Obtetivo:**

 Vamos deduzir o método dos Mínimos Quadrados Ordinários (MQO) para uma regressão linear simples.

 Dado um conjunto de pontos em um plano cartesiano $(x,y)$, vamos calcular uma reta:

 $$y = \beta_0 + \beta_1 x$$

 que passe o mais perto possível dos pontos dados.

 O objetivo é encontrar os coeficientes $\beta_0$ (intercepto) e $\beta_1$ (coeficiente angular) que minimizam a soma dos quadrados dos resíduos.

 O método será aplicado para obter os melhores coeficientes de forma a minimizar o erro da seguinte equação em relação aos valores tabelados da população brasileira:


$$P(t) = e^{t} \, e^{\alpha t}$$

onde $P(t)$ é a população brasileira em milhões de pessoas e $t$ é o tempo contado, em anos, a partir da data do primeiro censo em 1872.

Como esta equação não é linear, para a aplicação do método dos mínimos quadrados, faremos a seguinte transformação:

 $$ln P(t) = \alpha t + k$$

onde teremos a seguinte correspondência:

$y = ln P(t)$

$x = t$

$\beta_0 = k$

$\beta_1 = \alpha$



**Passo 1: Definição do Problema:**

Dado um conjunto de $n$ pontos $(x_i,y_i)$, queremos ajustar uma reta:

$$y = \beta_0 + \beta_1 x$$

Os resíduos (erros) são as diferenças entre os valores observados $y_i$ e os valores previstos $\hat{y_i} = \beta_0 + \beta_1 x_i$:

$$\epsilon_i = y_i - \hat{y_i} = y_i - \left( \beta_0 + \beta_1 x_i\right)$$

**Passo 2: Função Objetivo**:

Queremos minimizar a soma dos quadrados dos resíduos (SQE):

$$SQE = \sum_{i=1}^{n} \epsilon_i^2 = \sum_{i=1}^{n} \left( y_i - \beta_0 - \beta_1 x_i \right)^2$$

**Passo 3: Derivadas Parciais**:

Para minimizar SQE, derivamos em relação a $\beta_0$ e $\beta_1$ e igualamos a zero.


1.   Derivada em relação a $\beta_0$:

$$\dfrac{\partial SQE}{\partial \beta_0}= -2 \sum_{i=1}^{n} \left( y_i - \beta_0 - \beta_1 x_i \right) = 0$$

Simplificando:

$$\sum_{i=1}^{n} y_i - n \beta_0 - \beta_1 \sum_{i=1}^{n} x_i = 0$$

Rearranjando:

$$n \beta_0 + \beta_1 \sum_{i=1}^{n} x_i = \sum_{i=1}^{n} y_i$$  Equação 1


2.   Derivada em relação a $\beta_1$:



$$\dfrac{\partial SQE}{\partial \beta_1}= -2 \sum_{i=1}^{n} x_i \left( y_i - \beta_0 - \beta_1 x_i \right) = 0$$

Simplificando:

$$\sum_{i=1}^{n} x_i y_i - \beta_0 \sum_{i=1}^{n} x_i - \beta_1 \sum_{i=1}^{n} x_i^2  = 0$$

Rearranjando:

$$\beta_0 \sum_{i=1}^{n} x_i + \beta_1 \sum_{i=1}^{n} x_i^2 = \sum_{i=1}^{n} x_i y_i $$


**Passo 4: Sistema de Equações**:

Temos um sistema linear com duas equações e duas incógnitas $\left( \beta_0 \,\, e \,\, \beta_1 \right)$:

\begin{matrix}
n \beta_0 & + & \beta_1 \sum_{i=1}^{n} x_i & = & \sum_{i=1}^{n} y_i \\
\beta_0 \sum_{i=1}^{n} x_i & + & \beta_1 \sum_{i=1}^{n} x_i^2 & = & \sum_{i=1}^{n} x_i y_i
\end{matrix}

---

Escrevendo em forma de matriz:
$$
\begin{bmatrix}
n \beta_0 & \beta_1 \sum_{i=1}^{n} x_i \\
\beta_0 \sum_{i=1}^{n} x_i & \beta_1 \sum_{i=1}^{n} x_i^2
\end{bmatrix}
\begin{bmatrix}
\beta_0\\
\beta_1
\end{bmatrix}=
\begin{bmatrix}
\sum_{i=1}^{n} y_i\\
\sum_{i=1}^{n} x_i y_i
\end{bmatrix}
$$



**Passo 5: Solução do Sistema**:

1. Coeficiente angular $(\beta_1)$:

$$\beta_1 = \dfrac{n \sum_{i=1}^{n} x_i y_i - \sum_{i=1}^{n} x_i \sum_{i=1}^{n} y_i}{n \sum_{i=1}^{n} x_i^2 - \left( \sum_{i=1}^{n} x_i \right)^2}$$

2. Intercepto $( \beta_0)$:

$$\beta_0 = \dfrac{\sum_{i=1}^n y_i - \beta_1 \sum_{i=1}^n x_i }{n}$$

Como foi observado no gráfico abaixo, o modelo malthusiano leva a um crescimento exponencial da população. Porém esse pressuposto é falso como é possível observer no próprio gráfico.
